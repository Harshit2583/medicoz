```python
import pandas as pd   
import numpy as np 
```

```python
df = pd.read_csv('aasthma.csv')
df
```

```python
df.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in df.columns:
    if df[i].dtype == 'object':
        df[i] = le.fit_transform(df[i])
```

```python
df
```

```python
df
```

```python
df.nunique()
```

```python
for i in df.columns: 
    print(f" {df[i].unique()}")
```

```python
df['severity level'] = 0 

def severity_finder(row):
    if row['Severity_Mild'] == 1:
        return 2
    elif row['Severity_Moderate'] == 1:
        return 1
    elif row['Severity_None'] == 1:
        return 0
    return row['severity level']

df['severity level'] = df.apply(severity_finder, axis=1)
df
```

```python
df.drop(columns=['Severity_Mild', 'Severity_Moderate', 'Severity_None'], inplace=True)
df
```

```python
df.corr()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt


corr_matrix = df.corr()


plt.figure(figsize=(12, 10))


sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)


plt.show()
```
